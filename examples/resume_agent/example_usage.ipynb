# Interactive Resume Agent Example

This notebook demonstrates how to use the LangGraph-based Resume Agent for:
- Interactive resume creation
- Resume optimization
- Resume-JD matching with keyword highlighting
- Project experience ranking


```python
# Import required modules
from resume_agent import ResumeAgent, ResumeAgentState
from resume_models import Resume, ContactInfo, WorkExperience, JobDescription
from matching_tools import match_resume_with_jd, generate_resume_text
import json
```

## Example 1: Create a Sample Resume

Let's create a sample resume to demonstrate the functionality:

```python
# Create sample contact info
contact = ContactInfo(
    name="Alex Johnson",
    email="alex.johnson@email.com",
    phone="(555) 123-4567",
    location="San Francisco, CA",
    linkedin="https://linkedin.com/in/alexjohnson",
    github="https://github.com/alexjohnson"
)

# Create sample work experience
experience = WorkExperience(
    title="Senior Software Engineer",
    company="TechCorp Inc.",
    location="San Francisco, CA",
    start_date="2022-01",
    end_date="Present",
    description=[
        "Developed and maintained scalable web applications using Python and React",
        "Led a team of 4 engineers in implementing microservices architecture",
        "Improved application performance by 40% through code optimization",
        "Implemented CI/CD pipelines using Docker and Kubernetes"
    ],
    skills=["Python", "React", "Docker", "Kubernetes", "AWS"]
)

# Create sample resume
sample_resume = Resume(
    contact_info=contact,
    summary="Experienced software engineer with 5+ years developing scalable web applications and leading engineering teams.",
    work_experience=[experience],
    skills=["Python", "JavaScript", "React", "Node.js", "Docker", "Kubernetes", "AWS", "PostgreSQL", "MongoDB"]
)

print("Sample resume created!")
print(f"Contact: {sample_resume.contact_info.name}")
print(f"Skills: {', '.join(sample_resume.skills)}")
```

## Example 2: Generate Resume Text

Convert the resume to formatted text:

```python
# Generate formatted resume text
resume_text = generate_resume_text(sample_resume)
print("Generated Resume:")
print("=" * 50)
print(resume_text)
```

## Example 3: Job Description Matching

Create a job description and match it with the resume:

```python
# Create sample job description
job_desc = JobDescription(
    title="Full Stack Developer",
    company="StartupXYZ",
    location="Remote",
    description="We are looking for a full stack developer to join our growing team. The ideal candidate will have experience with modern web technologies and cloud platforms.",
    requirements=[
        "3+ years of experience with Python or JavaScript",
        "Experience with React or Vue.js",
        "Knowledge of database systems (PostgreSQL, MongoDB)",
        "Experience with cloud platforms (AWS, GCP, Azure)",
        "Understanding of containerization (Docker)"
    ],
    preferred_qualifications=[
        "Experience with Kubernetes",
        "Knowledge of microservices architecture",
        "Leadership experience",
        "CI/CD pipeline experience"
    ],
    responsibilities=[
        "Develop and maintain web applications",
        "Collaborate with cross-functional teams",
        "Implement scalable solutions",
        "Optimize application performance"
    ],
    keywords=["Python", "React", "AWS", "Docker", "Kubernetes", "PostgreSQL", "microservices"]
)

print("Job description created!")
print(f"Title: {job_desc.title} at {job_desc.company}")
print(f"Key requirements: {', '.join(job_desc.requirements[:3])}...")
```

## Example 4: Perform Resume-JD Matching

```python
# Perform matching
match_result = match_resume_with_jd(sample_resume, job_desc)

print("Resume-JD Matching Results:")
print("=" * 40)
print(f"Overall Match Score: {match_result.overall_score:.2%}")
print(f"\nMatching Keywords ({len(match_result.matching_keywords)}):")
print(", ".join(match_result.matching_keywords))
print(f"\nMissing Keywords ({len(match_result.missing_keywords)}):")
print(", ".join(match_result.missing_keywords))
print(f"\nRecommendations:")
for i, rec in enumerate(match_result.recommendations, 1):
    print(f"{i}. {rec}")
```

## Example 5: View Highlighted Resume

See the resume with matching keywords highlighted:

```python
# Display highlighted resume
print("Resume with Highlighted Keywords:")
print("=" * 50)
print(match_result.highlighted_resume)
```
